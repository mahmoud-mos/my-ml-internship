## 3. Quick look at the data (2-3 real numbers)

To justify our focus on stale content, we examined our starter dataset of 30,000 items. We found that a significant portion of our inventory is stale (defined as not updated in 180+ days), and a subset of those items still generate substantial impressions. This proves that an automated prioritization queue is necessary to help editors identify which of these stale assets deserve immediate attention.

```python
import pandas as pd

# Load the data
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

# Define the stale threshold and high-visibility threshold
stale_days = 180
min_impressions = 500

# Calculate the numbers
stale_assets = df[df['days_since_last_update'] >= stale_days]
stale_high_vis = stale_assets[stale_assets['impressions_90d'] >= min_impressions]

print(f"Total content items: {len(df)}")
print(f"Stale assets (>= {stale_days} days): {len(stale_assets)}")
print(f"High-visibility stale assets (>= {min_impressions} impressions): {len(stale_high_vis)}")
```
